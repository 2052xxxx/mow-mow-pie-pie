## R-3.1 
Graph the functions 8n, 4nlogn, 2n2, n3, and 2n using a logarithmic scale for the x- and y-axes; that is, if the function value f(n) is y, plot this as a point with x-coordinate at logn and y-coordinate at logy.

**Step 1.** For each function f(n), we need to plot the point (log(n), log(f(n))) on a graph with logarithmic scales for both x and y axes.

**Step 2.** Let's start with **8n**. Taking the logarithm of both sides, we get log(8n) = log(8) + log(n). 
            
            Therefore, the point on the graph will be **(log(n), log(8) + log(n))**.

**Step 3.** Next, we have **4nlog(n)**. Again, taking the logarithm of both sides, we get log(4nlog(n)) = log(4) + log(n) + log(log(n)).       
            
            Therefore, the point on the graph will be **(log(n), log(4) + log(n) + log(log(n)))**.

**Step 4.** Moving on to **2n^2**, we take the logarithm of both sides to get log(2n^2) = log(2) + 2log(n). 
            
            Therefore, the point on the graph will be **(log(n), log(2) + 2log(n))**.

**Step 5.** For **n^3**, we have log(n^3) = 3log(n). 
            
            Therefore, the point on the graph will be **(log(n), 3log(n))**.

**Step 6.** Finally, for **2^n**, taking the logarithm of both sides, we get log(2^n) = nlog(2). 
            
            Therefore, the point on the graph will be **(log(n), nlog(2))**.


```python
import matplotlib.pyplot as plt
import numpy as np
from math import log

n = np.array(range(2, 20))
x, f1, f2, f3, f4, f5 = [], [], [], [], [], []
functions = [f1, f2, f3, f4, f5]
labels = ["8n", "4nlogn", "2n²", "n³", "2ⁿ"]

for i in n:
    x.append(log(i))

    # (log(n), nlog(2)).
    f5.append(i*log(2))

for i in x:
    # (log(n), log(8) + log(n)).
    f1.append(log(8)+ i)

    # (log(n), log(4) + log(n) + log(log(n))).
    f2.append(log(4)+ i + log(i))

    # (log(n), log(2) + 2log(n)).
    f3.append(log(2)+ 2*i)

    # (log(n), 3log(n)).
    f4.append(3*i)

for i in range(5):
    plt.plot(x, functions[i], label=labels[i])

# Adding labels and legend
plt.xlabel('log(n)')
plt.ylabel('log(f(n))')
plt.legend()
plt.show()
```

## R-3.2
The number of operations executed by algorithms A and B is 8nlogn and 2n2, respectively. Determine n(0) such that A is better than B for n ≥ n(0)

```python
import math

# Function to find the smallest n0 such that 4log(n) < n
def find_n0():
    n = 1
    while True:
        if 4 * math.log(n) < n:
            return n
        n += 1

n0 = find_n0()
n0
```

## R-3.8 
Order the following functions by asymptotic growth rate:
\[
\begin{aligned}
4n \log n + 2n, & \quad 2^{10}, & \quad 2^{\log n}, \\
3n + 100\log n, & \quad 4n, & \quad 2^n, \\
n^2 + 10n, & \quad n^3, & \quad n \log n. \\
\end{aligned}
\]

### Ans

#### Phân tích các hàm:

1. **Hàm có dạng logarit, đa thức, và mũ:**
   - \( n \log n \): Hàm này tăng trưởng nhanh hơn \( n \) nhưng chậm hơn các hàm đa thức bậc cao hơn.
   - \( n^k \) (trong đó \( k \) là số nguyên dương): Tăng trưởng nhanh hơn so với \( n \log n \) và các hàm bậc thấp hơn. Cụ thể:
     - \( n \): tăng trưởng chậm nhất trong số các hàm đa thức.
     - \( n^2 \): tăng trưởng nhanh hơn \( n \) nhưng chậm hơn \( n^3 \).
     - \( n^3 \): tăng trưởng nhanh hơn \( n^2 \).
   - \( 2^n \): Hàm mũ tăng trưởng cực kỳ nhanh.
   - \( 2^{\log n} \): Tương đương với \( n \) vì \( 2^{\log n} = n^{\log 2} \), và \( \log 2 \) là hằng số.
   - \( 2^{10} \): Đây là hằng số (giá trị cố định), không phụ thuộc vào \( n \).

2. **Các hàm phức hợp:**
   - \( 4n \log n + 2n \): Thành phần \( 4n \log n \) là thành phần chính, nên hàm này tăng trưởng giống như \( n \log n \).
   - \( 3n + 100 \log n \): Thành phần \( 3n \) là thành phần chính, nên hàm này tăng trưởng giống như \( n \).
   - \( n^2 + 10n \): Thành phần \( n^2 \) là thành phần chính, nên hàm này tăng trưởng giống như \( n^2 \).

#### Sắp xếp các hàm:

1. **Hằng số:**
   - \( 2^{10} \) (đây là hằng số, không phụ thuộc vào \( n \)).

2. **Hàm tăng trưởng tuyến tính hoặc logarit:**
   - \( 3n + 100 \log n \) (tăng trưởng giống như \( n \)).
   - \( 4n \) (tăng trưởng giống như \( n \)).
   - \( 2^{\log n} \) (tăng trưởng giống như \( n \)).
  
3. **Hàm tăng trưởng nhanh hơn tuyến tính:**
   - \( n \log n \) (tăng trưởng nhanh hơn \( n \) nhưng chậm hơn \( n^2 \)).
   - \( 4n \log n + 2n \) (tăng trưởng giống như \( n \log n \)).
  
4. **Hàm bậc hai và cao hơn:**
   - \( n^2 + 10n \) (tăng trưởng giống như \( n^2 \)).
   - \( n^3 \) (tăng trưởng nhanh hơn \( n^2 \)).

5. **Hàm mũ:**
   - \( 2^n \) (tăng trưởng nhanh nhất trong các hàm được cho).

#### Sắp xếp cuối cùng từ chậm nhất đến nhanh nhất:

### Example 4.1.2: Drawing an English Ruler

```python
def draw_line(tick_length, tick_label=''):
    line = '-'* tick_length
    if tick_label:
        line += ' ' + tick_label
    print(line)

def draw_interval(center_length):
    if center_length > 0:
        draw_interval(center_length - 1)
        draw_line(center_length) 
        draw_interval(center_length - 1) 

def draw_ruler(num_inches, major_length):
    draw_line(major_length, '0' ) 
    for j in range(1, 1 + num_inches):
        draw_interval(major_length - 1) 
        draw_line(major_length, str(j))

draw_ruler(3, 3)
```
